```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
num_durations = 100  # Number of pulse duration points
num_amps = 100       # Number of drive amplitude points

duration_range = np.linspace(0, 500, num_durations)  # Pulse duration range (ns)
amplitude_range = np.linspace(0, 1, num_amps)  # Drive amplitude range (arb. units)

# Create 2D Rabi oscillation pattern
X, Y = np.meshgrid(duration_range, amplitude_range)
Rabi_frequency = Y * 10  # Assume Rabi frequency scales linearly with drive amplitude
Z = np.sin(np.pi * Rabi_frequency * X / 500) ** 2  # Excited state probability

# Plot the 2D colormap
plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, Z, cmap='RdBu', shading='auto')
plt.colorbar(label='Excited State Probability')
plt.xlabel('Pulse Duration (ns)')
plt.ylabel('Drive Amplitude')
plt.title('Rabi Oscillations in a Transmon Qubit')
plt.show()
```

```python
# Define parameters
num_durations = 100  # Number of pulse duration points
num_amps = 100       # Number of drive amplitude points
num_freqs = 100  # Number of frequency detuning points

```

```python
# Create 2D Rabi oscillation pattern - PULSE DURATION vs AMPLITUDE
X, Y = np.meshgrid(duration_range, amplitude_range)
Rabi_frequency = Y * 10  # Assume Rabi frequency scales linearly with drive amplitude
Z = np.sin(np.pi * Rabi_frequency * X / 500) ** 2  # Excited state probability

# Plot the 2D colormap
plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, Z, cmap='RdBu', shading='auto')
plt.colorbar(label='Excited State Probability')
plt.xlabel('Pulse Duration (ns)')
plt.ylabel('Drive Amplitude')
plt.title('Rabi Oscillations in a Transmon Qubit')
plt.show()
```

```python
duration_range = np.linspace(0, 500, num_durations)  # Pulse duration range (ns)
amplitude_range = np.linspace(0, 0.2, num_amps)  # Drive amplitude range (arb. units)
detuning_range = np.linspace(-0.25, 0.25, num_freqs)  # Frequency detuning range (MHz)

# Create 2D Rabi oscillation pattern - FREQUENCY DETINNING vs AMPLITUDE
t = 100 # pulse duration in ns
X, Y = np.meshgrid(detuning_range, amplitude_range)
Z = Y**2 /(Y**2 +X**2) * np.sin(np.sqrt(X**2+Y**2)*t/2) ** 2  # Excited state probability

# Plot the 2D colormap
sizeL=18
plt.figure(figsize=(8, 6))
plt.tick_params(axis='both', labelsize=14)
plt.pcolormesh(X, Y, Z, cmap='RdBu', shading='auto')
#plt.colorbar(label='Excited State Probability').ax.tick_params(labelsize=18)
cbar = plt.colorbar()
cbar.set_label('Excited State Probability', fontsize=sizeL)
cbar.ax.tick_params(labelsize=14)
plt.xlabel('Qubit frequency detuning (MHz)', size = sizeL)
plt.ylabel('Drive amplitude (A.U.)', size = sizeL)
#plt.title('Rabi Oscillations in a Transmon Qubit', size = sizeL)
plt.show()
```

```python
duration_range = np.linspace(0, 500, num_durations)  # Pulse duration range (ns)
amplitude_range = np.linspace(0, 10, num_amps)  # Drive amplitude range (arb. units)
detuning_range = np.linspace(-1, 1, num_freqs)  # Frequency detuning range (MHz)

# Create 2D Rabi oscillation pattern - FREQUENCY DETINNING vs AMPLITUDE
detuning = 10 # pulse duration in ns
X, Y = np.meshgrid(duration_range, amplitude_range)
Z = Y**2 /(Y**2 +detuning**2) *np.sin(np.sqrt(detuning**2+Y**2)*X/2) ** 2  # Excited state probability

# Plot the 2D colormap
plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, Z, cmap='RdBu', shading='auto')
plt.colorbar(label='Excited State Probability')
plt.xlabel('Pulse duration (ns)')
plt.ylabel('Drive amplitude (A.U.)')
plt.title('Rabi Oscillations in a Transmon Qubit')
plt.show()
```

```python
points = 601
detuningOverAmp = np.linspace(-2, 2, points)
ampTime = np.linspace(0, 10*np.pi, points)

# Create 2D Rabi oscillation pattern - FREQUENCY DETINNING vs AMPLITUDE
X, Y = np.meshgrid(ampTime, detuningOverAmp)
Z = 1 /(1+Y**2) * np.sin(np.sqrt(1+Y**2)*X/2) ** 2  # Excited state probability

# Plot the 2D colormap
sizeL=18
plt.figure(figsize=(8, 6))
plt.tick_params(axis='both', labelsize=14)
plt.pcolormesh(Y, X, Z, cmap='RdBu', shading='auto')
#plt.colorbar(label='Excited State Probability').ax.tick_params(labelsize=18)
cbar = plt.colorbar()
cbar.set_label('Excited State Probability', fontsize=sizeL)
cbar.ax.tick_params(labelsize=14)
plt.xlabel(r'Detuning / Amplitude ($\Delta / \Omega$)', size = sizeL)
plt.ylabel(r'Amplitude * Time ($\Omega t$)', size = sizeL)
#plt.title('Rabi Oscillations in a Transmon Qubit', size = sizeL)
plt.show()
```
