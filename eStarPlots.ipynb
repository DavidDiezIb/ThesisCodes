```python
import matplotlib.pyplot as plt
import numpy as np


eStarDataElecArgon ='/home/davidp/PlotsTesis/eStarData/StoppingPowerElectronsArgon.txt'
eStarDataElecNeon ='/home/davidp/PlotsTesis/eStarData/StoppingPowerElectronsNeon.txt'

# Load the data, skipping the first few header lines (adjust skip_header as needed)
data = np.genfromtxt(eStarDataElecArgon, delimiter=' ', skip_header=8)
dataN = np.genfromtxt(eStarDataElecNeon, delimiter=' ', skip_header=8)

# Extract each column into a separate array
energy = data[:, 0]
collision = data[:, 1]
radiative = data[:, 2]
total = data[:, 3]

energyN = dataN[:, 0]
collisionN = dataN[:, 1]
radiativeN = dataN[:, 2]
totalN = dataN[:, 3]
```

```python
# Configure Matplotlib to use LaTeX
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
})
```

```python
### STOPPING POWER ELECTRONS IN ARGON    ###

plt.figure(figsize=(10, 6))

plt.plot(energy, collision,  color = 'darkviolet', linestyle='dashed', label = 'Scattering', linewidth=3)
plt.plot(energy, radiative,  color = 'coral', linestyle='-.', label = 'Radiative', linewidth=3)
plt.plot(energy, total,  color = 'darkred', label = 'Total', linewidth=5)
#plt.plot(data_Ar1Iso['electric_field'], data_Ar1Iso['electron_drift_velocity'],  color = 'firebrick', label = r'Ar + $1\%$ $C_4H_{10}$')
#plt.plot(data_Ar2Iso['electric_field'], data_Ar2Iso['electron_drift_velocity'],  color = 'red', label = r'Ar + $2\%$ $C_4H_{10}$')
#plt.plot(data_Ar5Iso['electric_field'], data_Ar5Iso['electron_drift_velocity'],  color = 'tomato', label = r'Ar + $5\%$ $C_4H_{10}$')
#plt.plot(data_Ar10Iso['electric_field'], data_Ar10Iso['electron_drift_velocity'],  color = 'coral', label = r'Ar + $10\%$ $C_4H_{10}$')


plt.title(r'\textbf{Stopping power of electrons in argon}',  fontsize=25)
plt.xlabel(r'\textbf{Energy (MeV)}', fontsize=25)
plt.ylabel(r'\textbf{Stopping power (MeV$\,$cm$\,$/$\,$g)}', fontsize=25)
plt.xlim(1E-2, 1E3)
plt.ylim(1E-3, 1E2)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.xscale('log') 
plt.yscale('log')  
plt.legend(fontsize=25)
plt.grid(False)
#plt.savefig('StoppingPwrElecArgon.jpg', format='jpg', bbox_inches='tight')
#plt.savefig('StoppingPwrElecArgon.png', format='png', bbox_inches='tight')
#plt.savefig('StoppingPwrElecArgon.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
### STOPPING POWER ELECTRONS IN NEON    ###

plt.figure(figsize=(15, 10))

plt.plot(energyN, collisionN,  color = 'darkcyan', linestyle='dashed', label = 'Collision')
plt.plot(energyN, radiativeN,  color = 'olive', linestyle='dashed', label = 'Radiative')
plt.plot(energyN, totalN,  color = 'darkgreen', label = 'Total', linewidth=2)
#plt.plot(data_Ar1Iso['electric_field'], data_Ar1Iso['electron_drift_velocity'],  color = 'firebrick', label = r'Ar + $1\%$ $C_4H_{10}$')
#plt.plot(data_Ar2Iso['electric_field'], data_Ar2Iso['electron_drift_velocity'],  color = 'red', label = r'Ar + $2\%$ $C_4H_{10}$')
#plt.plot(data_Ar5Iso['electric_field'], data_Ar5Iso['electron_drift_velocity'],  color = 'tomato', label = r'Ar + $5\%$ $C_4H_{10}$')
#plt.plot(data_Ar10Iso['electric_field'], data_Ar10Iso['electron_drift_velocity'],  color = 'coral', label = r'Ar + $10\%$ $C_4H_{10}$')


plt.title(r'\textbf{Stopping power of electrons in neon}',  fontsize=35)
plt.xlabel(r'\textbf{Energy (MeV)}', fontsize=25)
plt.ylabel(r'\textbf{Stopping power (MeV cm / g)}', fontsize=25)
plt.xlim(1E-2, 1E3)
plt.ylim(1E-3, 1E2)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.xscale('log') 
plt.yscale('log')  
plt.legend(fontsize=25)
plt.grid(False)
#plt.savefig('StoppingPwrElecNeon.jpg', format='jpg', bbox_inches='tight')
#plt.savefig('StoppingPwrElecNeon.png', format='png', bbox_inches='tight')
#plt.savefig('StoppingPwrElecNeon.svg', format='svg', bbox_inches='tight')
plt.show()
```
