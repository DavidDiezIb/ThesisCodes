```python
import json
import matplotlib.pyplot as plt
import numpy as np

Ar1Iso = './GasDataFiles/Ar_99.0-C4H10_1.0.gas.json'
Ar2Iso = './GasDataFiles/Ar_98.0-C4H10_2.0.gas.json'
Ar5Iso = './GasDataFiles/Ar_95.0-C4H10_5.0.gas.json'
Ar10Iso = './GasDataFiles/Ar_90.0-C4H10_10.0.gas.json'
Ar = './GasDataFiles/Ar.gas.json'

Ne1Iso = './GasDataFiles/Ne_99.0-C4H10_1.0.gas.json'
Ne2Iso = './GasDataFiles/Ne_98.0-C4H10_2.0.gas.json'
Ne5Iso = './GasDataFiles/Ne_95.0-C4H10_5.0.gas.json'
Ne10Iso = './GasDataFiles/Ne_90.0-C4H10_10.0.gas.json'
Ne = './GasDataFiles/Ne.gas.json'


with open(Ar1Iso, 'r') as f:
    data_Ar1Iso = json.load(f)

with open(Ar2Iso, 'r') as f:
    data_Ar2Iso = json.load(f)

with open(Ar5Iso, 'r') as f:
    data_Ar5Iso = json.load(f)

with open(Ar10Iso, 'r') as f:
    data_Ar10Iso = json.load(f)

with open(Ar, 'r') as f:
    data_Ar = json.load(f)

with open(Ne1Iso, 'r') as f:
    data_Ne1Iso = json.load(f)

with open(Ne2Iso, 'r') as f:
    data_Ne2Iso = json.load(f)

with open(Ne5Iso, 'r') as f:
    data_Ne5Iso = json.load(f)

with open(Ne10Iso, 'r') as f:
    data_Ne10Iso = json.load(f)

with open(Ne, 'r') as f:
    data_Ne = json.load(f)
```

```python
# Configure Matplotlib to use LaTeX
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
})
```

```python
### ARGON DRIFT VELOICITY ###

plt.figure(figsize=(10, 6))

plt.plot(data_Ar['electric_field'], data_Ar['electron_drift_velocity'],  color = 'darkred', label = 'Ar')
plt.plot(data_Ar1Iso['electric_field'], data_Ar1Iso['electron_drift_velocity'],  color = 'firebrick', label = r'Ar + $1\%$ $C_4H_{10}$')
plt.plot(data_Ar2Iso['electric_field'], data_Ar2Iso['electron_drift_velocity'],  color = 'red', label = r'Ar + $2\%$ $C_4H_{10}$')
plt.plot(data_Ar5Iso['electric_field'], data_Ar5Iso['electron_drift_velocity'],  color = 'tomato', label = r'Ar + $5\%$ $C_4H_{10}$')
plt.plot(data_Ar10Iso['electric_field'], data_Ar10Iso['electron_drift_velocity'],  color = 'coral', label = r'Ar + $10\%$ $C_4H_{10}$')


plt.title(r'\textbf{Electron drift velocity}',  fontsize=25)
plt.xlabel(r'\textbf{Electric field / Pressure (V/cm/bar)}', fontsize=25)
plt.ylabel(r'\textbf{Drift velocity (cm/$\mu$s)}', fontsize=25)
plt.xlim(0, 10000)
plt.ylim(0, 5)
plt.tick_params(axis='both', which='major', labelsize=25)
#plt.yscale('log')  
plt.legend(fontsize=20)
plt.grid(False)
#plt.savefig('DriftVelocityArgon.jpg', format='jpg', bbox_inches='tight')
#plt.savefig('DriftVelocityArgon.png', format='png', bbox_inches='tight')
#plt.savefig('DriftVelocityArgon.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
### NEON DRIFT VELOICITY ###

plt.figure(figsize=(15, 10))

plt.plot(data_Ne['electric_field'], data_Ne['electron_drift_velocity'],  color = 'darkgreen', label = 'Ne')
plt.plot(data_Ne1Iso['electric_field'], data_Ne1Iso['electron_drift_velocity'],  color = 'green', label = r'Ne + $1\%$ $C_4H_{10}$')
plt.plot(data_Ne2Iso['electric_field'], data_Ne2Iso['electron_drift_velocity'],  color = 'limegreen', label = r'Ne + $2\%$ $C_4H_{10}$')
plt.plot(data_Ne5Iso['electric_field'], data_Ne5Iso['electron_drift_velocity'],  color = 'lime', label = r'Ne + $5\%$ $C_4H_{10}$')
plt.plot(data_Ne10Iso['electric_field'], data_Ne10Iso['electron_drift_velocity'],  color = 'mediumspringgreen', label = r'Ne + $10\%$ $C_4H_{10}$')


plt.title(r'\textbf{Drift velocity}',  fontsize=35)
plt.xlabel(r'\textbf{Electric field / Pressure (V/cm/bar)}', fontsize=25)
plt.ylabel(r'\textbf{Drift velocity (cm/$\mu$s)}', fontsize=25)
plt.xlim(0, 10000)
plt.ylim(0, 10)
plt.tick_params(axis='both', which='major', labelsize=25)
#plt.yscale('log')  
plt.legend(fontsize=25)
plt.grid(False)
#plt.savefig('DriftVelocityNeon.jpg', format='jpg', bbox_inches='tight')
#plt.savefig('DriftVelocityNeon.png', format='png', bbox_inches='tight')
#plt.savefig('DriftVelocityNeon.svg', format='svg', bbox_inches='tight')
plt.show()
```
